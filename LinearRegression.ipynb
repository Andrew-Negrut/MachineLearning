```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

```python
# hypothesis
def h_theta(theta, x):
  return np.dot(theta, x)
```

```python
def cost(theta, x, m):
  res = 0
  for i in range(m):
    res += (h_theta(theta, x[i]) - y[i]) ** 2

  return res / 2.0
```

```python
def g_descent(theta, x, y, m, n, alpha):
  for trial in range(1000):
    for i in range(m):
      theta_new = np.zeros(n)

      for j in range(n):
        theta_new[j] = theta[j] - alpha * x[i, j] * (h_theta(theta, x[i, :]) - y[i])

      theta = theta_new.copy()

    print("Trial ", trial, " cost: ", cost(theta, x, m))


  return theta
```

```python
# Generate larger dataset
x, y = make_regression(n_samples=100, n_features=1, noise=10)
plt.scatter(x, y)
plt.show()
```

```python
# adding in x_0 = 1

x = np.hstack((np.ones((x.shape[0], 1)), x))
```

```python
theta = np.zeros(x.shape[1])
```

```python
theta_final = g_descent(theta, x, y, x.shape[0], x.shape[1], 0.001)
theta_final
```

```python
plt.scatter(x[:, 1], y)

plt.axhline(0, color='black', linewidth=.5)
plt.axvline(0, color='black', linewidth=.5)

x1_inp = np.linspace(min(x[:, 1]), max(x[:, 1]))
plt.plot(x1_inp, theta_final[0] + theta_final[1] * x1_inp, linestyle="dashed", linewidth=2, color='g')
plt.show()
```
